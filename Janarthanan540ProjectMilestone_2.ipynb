**Step #5 introduce a new column derive based on 'Opinion' column value**


- Convert the confidence percentage into a number rating betweeen 1 to 10)
- For Buy confidence signal convert it to a psotive integer between  1 to 10  and 
- for Sell signal  convert it to a negative integer ranging between 1 to 10  
- Create a method to recieve the  stirng value and split them and identify the 'Buy'  or'Sell'  value and use it to denote positive or negative rating
-  use the magnitude value to convert % to 1 to 10 range value

```python
import math

# Declare a method to process the input string and return a postive or negative rating value from 1 to 10
def convertOpinionBuySellToratingScale(x):
    if x == "" or  type(x) == float or  x.find('%') == -1:
        return ""
    else:
        split_val = x.split('%')
        pos_neg_ind = -1 if split_val[1] == " Sell" else 1
        return  round(int(split_val[0])/10)*pos_neg_ind
    
```

```python
# create a new column to set the derived values
sp500_df['BuySell_rating'] = [convertOpinionBuySellToratingScale(x) for x in sp500_df['Opinion']]
sp500_df[['Opinion', 'BuySell_rating']].head(10)
```
